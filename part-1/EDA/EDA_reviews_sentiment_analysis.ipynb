## Sentiment Analysis EDA - Dec 20 reviews

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from datetime import datetime
from nltk.sentiment.vader import SentimentIntensityAnalyzer

%config InlineBackend.figure_format = 'retina'
%matplotlib inline

sns.set_theme(context='notebook', style='darkgrid', palette='deep', font_scale=1.5)
```

### Load data

```python
# load data for Dec 20 reviews with sentiment analysis
df = pd.read_csv('../GA/capstone-data-airbnb/dec20-data/reviews_sentiment_analysis.csv', index_col=0)
df.shape
```

```python
df_price = pd.read_csv('../GA/capstone-data-airbnb/dec20-data/active_listings_clean.csv', index_col=0)
df_price.shape
```

```python
# convert excel date serial numbers to dates
df['date'] = pd.to_datetime(df['date'], unit='D', origin='1899-12-30')
```

### Sentiment Analysis EDA

```python
# add review word count
df['review_length'] = df.comments.map(lambda x: len(x.split()))
```

```python
sid = SentimentIntensityAnalyzer()
# check sentiment analysis scores on an individual review
i = 1178605
print(df.comments[i])
ss = sid.polarity_scores(df.comments[i])
for k in sorted(ss):
    print('{0}: {1}, '.format(k, ss[k]), end='')
```

```python
# plot review lengths
fig, ax = plt.subplots(figsize=(14, 8))
sns.histplot(df, x='review_length', ax=ax, bins=200)
ax.set_xlabel('Number of words in review')
ax.set_ylabel('Count')
ax.set_title('Distribution of review lengths')
plt.show()
```

```python
# plot compound polarity scores
fig, ax = plt.subplots(figsize=(14, 8))
sns.histplot(df, x='sa_compound', ax=ax, bins=200)
ax.set_xlabel('Compound polarity score')
ax.set_ylabel('Count')
ax.set_title('Distribution of VADER sentiment analysis compound polarity scores')
plt.show()
```

```python
# plot distributions of each score
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(16, 12))
sns.histplot(df.sa_neg, ax=ax[0][0], bins=100)
sns.histplot(df.sa_neu, ax=ax[0][1], bins=100)
sns.histplot(df.sa_pos, ax=ax[1][0], bins=100)
sns.histplot(df.sa_compound, ax=ax[1][1], bins=100)

for a in ax:
    for z in a:
        z.set_xlabel(None)
        
ax[0][0].set_title('Negative score')
ax[0][1].set_title('Neutral score')
ax[1][0].set_title('Positive score')
ax[1][1].set_title('Compound score')

plt.show()
```

```python
# plot compound polarity scores versus review length
fig, ax = plt.subplots(figsize=(16, 8))
sns.scatterplot(data=df, x='sa_compound', y='review_length', ax=ax)
ax.set_xlabel('Compound polarity score')
ax.set_ylabel('Number of words in review')
ax.set_title('Review length versus compound score')
plt.show()
```

### Investigate relationship with price

```python
price = df_price[['id', 'price']].copy()
price.rename(columns={'id': 'listing_id'}, inplace=True)
# merge listing prices
df_plot = pd.merge(df, price, how='left', on='listing_id')
df_plot.dropna(inplace=True)
df_plot.shape
```

```python
# plot each score versus price
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(16, 12))
sns.scatterplot(x=df_plot.sa_neg, y=df_plot.price, ax=ax[0][0])
sns.scatterplot(x=df_plot.sa_neu, y=df_plot.price, ax=ax[0][1])
sns.scatterplot(x=df_plot.sa_pos, y=df_plot.price, ax=ax[1][0])
sns.scatterplot(x=df_plot.sa_compound, y=df_plot.price, ax=ax[1][1])

for a in ax:
    for z in a:
        z.set_xlabel(None)
        
ax[0][0].set_title('Negative score')
ax[0][1].set_title('Neutral score')
ax[1][0].set_title('Positive score')
ax[1][1].set_title('Compound score')

plt.show()
```

```python
# calculate pairwise correlations between sentiment analysis scores and price
corr = df_plot[['sa_compound', 'sa_neg', 'sa_neu', 'sa_pos', 'review_length', 'price']].corr()

fig, ax = plt.subplots(figsize=(8, 6))

# generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# plot the heatmap with seaborn
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', ax=ax, cmap='vlag')
ax.set_ylim([corr.shape[1], 0])

        
plt.title('Correlation between numerical features and price')
plt.show()
```

```python
# aggregate by listing id
df_plot_agg = df.groupby('listing_id').sa_compound.agg(['mean', 'median', 'count', 'max', 'min']).reset_index()
df_plot_agg = pd.merge(price, df_plot_agg, how='left', on='listing_id')
df_plot_agg.shape
```

```python
# plot aggregated compound polarity scores distribution and versus price
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font_scale=1.3)
fig, ax = plt.subplots(ncols=2, figsize=(18, 7))

sns.histplot(df_plot_agg, x='mean', ax=ax[0], bins=200)
ax[0].set_xlabel('Compound polarity score')
ax[0].set_ylabel('Count')
ax[0].set_title('Distribution of VADER sentiment analysis compound scores')

sns.scatterplot(x=df_plot_agg['mean'], y=df_plot_agg.price, ax=ax[1])
ax[1].set_xlabel('Compound polarity score')
ax[1].set_ylabel('Listing price')
ax[1].set_title('VADER sentiment analysis compound scores versus price')
plt.show()
```

### Aggregating reviews

```python
df_agg = df.groupby('listing_id').sa_compound.agg(['mean', 'median', 'std', 'skew', stats.kurtosis, 'count', 'max', 'min']).reset_index()
df_agg.shape
```

```python
# write aggregated data to CSV
df_agg.to_csv('../capstone-data-airbnb/dec20-data/reviews_sentiment_analysis_aggregated.csv')
```
