## Part 1 - Predicting Airbnb rental prices

## Modelling with XGBoost

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
import xgboost as xgb

%config InlineBackend.figure_format = 'retina'
%matplotlib inline

sns.set_theme(context='notebook', style='darkgrid', palette='deep', font_scale=1.5)
```

```python
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PowerTransformer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
from sklearn.compose import TransformedTargetRegressor
```

```python
# load data
df = pd.read_csv('../GA/capstone-data-airbnb/dec20-data/active_listings_modelling.csv', index_col=0)

# replace null values in text features with empty strings to use with CountVectorizer
df.name.replace(np.nan, '', inplace=True)
df.description.replace(np.nan, '', inplace=True)
df.neighborhood_overview.replace(np.nan, '', inplace=True)
df.shape
```

```python
categorical = ['host_response_time', 'host_response_rate', 'host_acceptance_rate',
               'neighbourhood_cleansed', 'room_type']

numerical = ['host_is_superhost', 'host_listings_count', 'host_identity_verified',
             'accommodates', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 
             'availability_30', 'availability_90', 'availability_365', 
             'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d', 
             'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
             'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 
             'review_scores_value', 'first_review_days', 'last_review_days', 'instant_bookable', 
             'reviews_per_month', 'number_bathrooms', 'private_bathroom', 'number_beds', 
             'number_bedrooms', 'host_since_days', 'sentiment_analysis_mean', 
             'sentiment_analysis_std', 'sentiment_analysis_skew', 'sentiment_analysis_kurtosis',  
             'British Museum', 'Tate Modern', 'Natural History Museum', 'Southbank Centre', 
             'Tower of London', 'Royal Museums Greenwich', 'Kew Gardens', 'Tate Britain']

text = ['name', 'description', 'property_type', 'amenities']

unused = ['id', 'host_name', 'neighborhood_overview', 'V&A Museum', 'Science Museum', 
          'Somerset House', 'National Gallery', 'availability_60', 'sentiment_analysis_median', 
          'sentiment_analysis_count', 'sentiment_analysis_max', 'sentiment_analysis_min']

len(categorical) + len(numerical) + len(text) + len(unused)
```

### Train-test split and pipeline

```python
y = df.price.copy()
X = df[categorical + numerical + text].copy()

# do train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# create custom stop words for NLP
from sklearn.feature_extraction import text
property_stop_words = ['private', 'room', 'entire', 'in', 'and', 'shared', 'bed', 'apt', 'guest', 'place']
additional_words = ['br']
description_stop_words = text.ENGLISH_STOP_WORDS.union(additional_words)
```

```python
# create pipepline with additional features
one_hot = OneHotEncoder(sparse=False, drop='first')

cvec_amenities = CountVectorizer(token_pattern='(?u)[a-zA-Z][a-z ]+',
                                 ngram_range=(1,1),
                                 min_df=5)

cvec_property_type = CountVectorizer(stop_words=property_stop_words,
                                     ngram_range=(1,1),
                                     min_df=5)

tvec_name = TfidfVectorizer(stop_words='english',
                            ngram_range=(1,2),
                            max_features=None,
                            min_df=10)

tvec_description = TfidfVectorizer(stop_words=description_stop_words,
                                   ngram_range=(2,2),
                                   max_features=500)

col_trans = ColumnTransformer([('amenities', cvec_amenities, 'amenities'),
                               ('property_type', cvec_property_type, 'property_type'),
                               ('name', tvec_name, 'name'),
                               ('description', tvec_description, 'description'),
                               ('categories', one_hot, categorical)],  
                               remainder='passthrough', 
                               sparse_threshold=0.6) 

pipe = Pipeline(steps=[('col_trans', col_trans),
                       ('scaler', StandardScaler(with_mean=False))])
```

```python
X_train = pipe.fit_transform(X_train)
X_test = pipe.transform(X_test)
```

### Linear regression model

#### Without power transform

```python
lr = LinearRegression(n_jobs=-2)
lr.fit(X_train, y_train)
lr_scores = cross_val_score(lr, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', lr_scores.mean())
print('Training score: ', lr.score(X_train, y_train))
print('Test score: ', lr.score(X_test, y_test))
```

### Gradient boosting

```python
gradient_boosting = GradientBoostingRegressor(loss='ls',
                                              validation_fraction=0.1,
                                              n_iter_no_change=20,
                                              random_state=1)

params = {'n_estimators': [500, 700, 900],
          'max_depth': [5, 6, 7, 8],
          'learning_rate': [0.001, 0.01, 0.1]}

gradient_boosting_gs = GridSearchCV(gradient_boosting,
                                    params,
                                    n_jobs=-2, 
                                    cv=5, 
                                    verbose=3)

gradient_boosting_gs.fit(X_train, y_train)
```

```python
print('Best parameters:')
print(gradient_boosting_gs.best_params_)
print('Best estimator mean cross-validated training score:')
print(gradient_boosting_gs.best_score_)
print('Best estimator score on the full training set:')
print(gradient_boosting_gs.score(X_train, y_train))
print('Best estimator score on the test set:')
print(gradient_boosting_gs.score(X_test, y_test))
```

```python
gb_gs_best = gradient_boosting_gs.best_estimator_
```

```python
joblib.dump(gb_gs_best, '../capstone-models/gb_model_gs.jlib')
```

```python
print('MAE on training set: ', mean_absolute_error(y_train, gb_gs_best.predict(X_train)))
print('MAE on test set: ', mean_absolute_error(y_test, gb_gs_best.predict(X_test)))
```

### XGBoost model

#### With power transform

```python
model_pt = TransformedTargetRegressor(regressor=XGBRegressor(objective='reg:squarederror', n_estimators=200),
                                      transformer=PowerTransformer(method='box-cox'))
model_pt.fit(X_train, y_train)
model_pt_scores = cross_val_score(model_pt, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', model_pt_scores.mean())
print('Training score: ', model_pt.score(X_train, y_train))
print('Test score: ', model_pt.score(X_test, y_test))
```

#### Without power transform

```python
model = XGBRegressor(objective='reg:squarederror', n_estimators=600)
model.fit(X_train, y_train)
model_scores = cross_val_score(model, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', model_scores.mean())
print('Training score: ', model.score(X_train, y_train))
print('Test score: ', model.score(X_test, y_test))
```

#### Gridsearch

```python
xgb = XGBRegressor(objective='reg:squarederror', random_state=1)

params = {'n_estimators': [500, 600],
          'max_depth': [7, 8, 9],
          'min_child_weight': [6, 7, 8],
          'learning_rate': [0.01, 0.05, 0.1]}

xgb_gs = GridSearchCV(xgb,
                      params,
                      n_jobs=-2, 
                      cv=5, 
                      verbose=3)

xgb_gs.fit(X_train, y_train)
```

```python
print('Best parameters:')
print(xgb_gs.best_params_)
print('Best estimator mean cross-validated training score:')
print(xgb_gs.best_score_)
print('Best estimator score on the full training set:')
print(xgb_gs.score(X_train, y_train))
print('Best estimator score on the test set:')
print(xgb_gs.score(X_test, y_test))
```

```python
xgb_gs_best = xgb_gs.best_estimator_
```

```python
print('MAE on training set: ', mean_absolute_error(y_train, xgb_gs_best.predict(X_train)))
print('MAE on test set: ', mean_absolute_error(y_test, xgb_gs_best.predict(X_test)))
```

```python
joblib.dump(xgb_gs_best, '../capstone-models/xgb_model_gs.jlib')
```

```python
# plot feature importances xgboost
model = xgb_gs_best
df_coef = pd.DataFrame({'variable': col_trans.get_feature_names(), 
                        'coef': model.feature_importances_})

fig4, ax = plt.subplots(figsize=(10,10))
sns.barplot(data=df_coef.sort_values('coef', ascending=False).head(20), x='coef', y='variable')
plt.title('Feature importances - top 20', fontsize=20)
plt.ylabel(None)
plt.show()
```

```python
# plot feature importances random forest
model = rf_reloaded
df_coef = pd.DataFrame({'variable': col_trans.get_feature_names(), 
                        'coef': model.feature_importances_})

fig4, ax = plt.subplots(figsize=(10,10))
sns.barplot(data=df_coef.sort_values('coef', ascending=False).head(20), x='coef', y='variable')
plt.title('Feature importances - top 20', fontsize=20)
plt.ylabel(None)
plt.show()
```

```python
# plot true versus predicted values
plt.figure(figsize=(14, 10))
plt.scatter(xgb_gs_best.predict(X_test), y_test, alpha=0.5)
plt.scatter(rf_reloaded.predict(X_test), y_test, alpha=0.5)
plt.legend(['XGBoost', 'Random Forest'])
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], c='k', lw=2)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Predicted Values versus Actual Values')
plt.show()
```

```python
# plot the residuals
fig, ax = plt.subplots(ncols=2, figsize=(16, 6))
sns.histplot(y_test - xgb_gs_best.predict(X_test), ax=ax[0])
ax[0].set_title('XGBoost')
sns.histplot(y_test - rf_reloaded.predict(X_test), ax=ax[1])
ax[1].set_title('Random Forest')

for a in ax:
    a.set_xlabel('Residuals')
    a.set_ylabel('Count')

plt.show()
```
