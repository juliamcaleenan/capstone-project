## Modelling with additional engineered features

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import joblib

%config InlineBackend.figure_format = 'retina'
%matplotlib inline

sns.set_theme(context='notebook', style='darkgrid', palette='deep', font_scale=1.5)
```

```python
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PowerTransformer, PolynomialFeatures
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV, ElasticNet, ElasticNetCV
from sklearn.ensemble import RandomForestRegressor
from sklearn import svm
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.svm import SVR, LinearSVR
from sklearn.compose import TransformedTargetRegressor
from sklearn.neural_network import MLPRegressor
```

```python
# load data
df = pd.read_csv('../GA/capstone-data-airbnb/dec20-data/active_listings_modelling.csv', index_col=0)

# replace null values in text features with empty strings to use with CountVectorizer
df.name.replace(np.nan, '', inplace=True)
df.description.replace(np.nan, '', inplace=True)
df.neighborhood_overview.replace(np.nan, '', inplace=True)
df.shape
```

```python
categorical = ['host_response_time', 'host_response_rate', 'host_acceptance_rate',
               'neighbourhood_cleansed', 'room_type']

numerical = ['host_is_superhost', 'host_listings_count', 'host_identity_verified',
             'accommodates', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 
             'availability_30', 'availability_90', 'availability_365', 
             'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d', 
             'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
             'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 
             'review_scores_value', 'first_review_days', 'last_review_days', 'instant_bookable', 
             'reviews_per_month', 'number_bathrooms', 'private_bathroom', 'number_beds', 
             'number_bedrooms', 'host_since_days', 'sentiment_analysis_mean', 
             'sentiment_analysis_std', 'sentiment_analysis_skew', 'sentiment_analysis_kurtosis',  
             'British Museum', 'Tate Modern', 'Natural History Museum', 'Southbank Centre', 
             'Tower of London', 'Royal Museums Greenwich', 'Kew Gardens', 'Tate Britain']

text = ['name', 'description', 'property_type', 'amenities']

unused = ['id', 'host_name', 'neighborhood_overview', 'V&A Museum', 'Science Museum', 
          'Somerset House', 'National Gallery', 'availability_60', 'sentiment_analysis_median', 
          'sentiment_analysis_count', 'sentiment_analysis_max', 'sentiment_analysis_min']

len(categorical) + len(numerical) + len(text) + len(unused)
```

### Train-test split and pipeline

```python
y = df.price.copy()
X = df[categorical + numerical + text].copy()

# do train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# create custom stop words for NLP
from sklearn.feature_extraction import text
property_stop_words = ['private', 'room', 'entire', 'in', 'and', 'shared', 'bed', 'apt', 'guest', 'place']
additional_words = ['br']
description_stop_words = text.ENGLISH_STOP_WORDS.union(additional_words)
```

```python
# create pipepline with additional features
one_hot = OneHotEncoder(sparse=False, drop='first')

cvec_amenities = CountVectorizer(token_pattern='(?u)[a-zA-Z][a-z ]+',
                                 ngram_range=(1,1),
                                 min_df=5)

cvec_property_type = CountVectorizer(stop_words=property_stop_words,
                                     ngram_range=(1,1),
                                     min_df=5)

tvec_name = TfidfVectorizer(stop_words='english',
                            ngram_range=(1,2),
                            max_features=None,
                            min_df=10)

tvec_description = TfidfVectorizer(stop_words=description_stop_words,
                                   ngram_range=(2,2),
                                   max_features=500)

polynomial = PolynomialFeatures(degree=2)

col_trans = ColumnTransformer([('amenities', cvec_amenities, 'amenities'),
                               ('property_type', cvec_property_type, 'property_type'),
                               ('name', tvec_name, 'name'),
                               ('description', tvec_description, 'description'),
#                                ('polynomial', polynomial, ['accommodates', 'number_bedrooms', 'number_bathrooms']),
                               ('categories', one_hot, categorical)],  
                               remainder='passthrough', 
                               sparse_threshold=0.6) 

pipe = Pipeline(steps=[('col_trans', col_trans),
                       ('scaler', StandardScaler(with_mean=False))])
```

```python
from sklearn import set_config
set_config(display='diagram') 
pipe
```

```python
X_train = pipe.fit_transform(X_train)
X_test = pipe.transform(X_test)
```

### Linear regression model

#### With power transform

```python
lr_pt = TransformedTargetRegressor(regressor=LinearRegression(),
                                   transformer=PowerTransformer(method='box-cox'))
lr_pt.fit(X_train, y_train)
lr_pt_scores = cross_val_score(lr_pt, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', lr_pt_scores.mean())
print('Training score: ', lr_pt.score(X_train, y_train))
print('Test score: ', lr_pt.score(X_test, y_test))
```

#### Without power transform

```python
lr = LinearRegression(n_jobs=-2)
lr.fit(X_train, y_train)
lr_scores = cross_val_score(lr, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', lr_scores.mean())
print('Training score: ', lr.score(X_train, y_train))
print('Test score: ', lr.score(X_test, y_test))
```

### Ridge model

#### With power transform

```python
# find optimal value for alpha in Ridge model
optimal_ridge_pt = TransformedTargetRegressor(regressor=RidgeCV(alphas=np.logspace(0, 5, 200), cv=5),
                                              transformer=PowerTransformer(method='box-cox'))
optimal_ridge_pt.fit(X_train, y_train)
print(optimal_ridge_pt.regressor_.alpha_)
```

```python
# look at cross-validation scores with this Ridge model
ridge_pt = TransformedTargetRegressor(regressor=Ridge(alpha=optimal_ridge_pt.regressor_.alpha_),
                                      transformer=PowerTransformer(method='box-cox'))
ridge_pt.fit(X_train, y_train)
ridge_pt_scores = cross_val_score(ridge_pt, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', ridge_pt_scores.mean())
print('Training score: ', ridge_pt.score(X_train, y_train))
print('Test score: ', ridge_pt.score(X_test, y_test))
```

#### Without power transform

```python
# find optimal value for alpha in Ridge model
optimal_ridge = RidgeCV(alphas=np.logspace(0, 5, 200), cv=5)
optimal_ridge.fit(X_train, y_train)
print(optimal_ridge.alpha_)
```

```python
# look at cross-validation scores with this Ridge model
ridge = Ridge(alpha=optimal_ridge.alpha_)
ridge.fit(X_train, y_train)
ridge_scores = cross_val_score(ridge, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', ridge_scores.mean())
print('Training score: ', ridge.score(X_train, y_train))
print('Test score: ', ridge.score(X_test, y_test))
```

```python
print('MAE on training set: ', mean_absolute_error(y_train, ridge.predict(X_train)))
print('MAE on test set: ', mean_absolute_error(y_test, ridge.predict(X_test)))
```

#### Model coefficients

```python
# collect model coefficients in a DataFrame
model = ridge
df_coef = pd.DataFrame({'variable': col_trans.get_feature_names(), 
                        'coef': model.coef_, 'coef_abs': np.abs(model.coef_)})
# plot coefficients from model
fig, ax = plt.subplots(figsize=(14,8))
sns.barplot(data=df_coef.sort_values('coef_abs', ascending=False).head(20), x='coef', y='variable')
plt.title('The 20 coefficients from the model with the most impact')
plt.show()
```

### Lasso model

#### With power transform

```python
# find optimal value for alpha in Lasso model
optimal_lasso_pt = TransformedTargetRegressor(regressor=LassoCV(n_alphas=200, cv=5, n_jobs=-2),
                                              transformer=PowerTransformer(method='box-cox'))
optimal_lasso_pt.fit(X_train, y_train)
print(optimal_lasso_pt.regressor_.alpha_)
```

```python
# look at cross-validation scores with this Lasso model
lasso_pt = TransformedTargetRegressor(regressor=Lasso(alpha=optimal_lasso_pt.regressor_.alpha_),
                                      transformer=PowerTransformer(method='box-cox'))
lasso_pt.fit(X_train, y_train)
lasso_pt_scores = cross_val_score(lasso_pt, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', lasso_pt_scores.mean())
print('Training score: ', lasso_pt.score(X_train, y_train))
print('Test score: ', lasso_pt.score(X_test, y_test))
```

```python
print('MAE on training set: ', mean_absolute_error(y_train, lasso_pt.predict(X_train)))
print('MAE on test set: ', mean_absolute_error(y_test, lasso_pt.predict(X_test)))
```

#### Without power transform

```python
# find optimal value for alpha in Lasso model
optimal_lasso = LassoCV(n_alphas=200, cv=5, max_iter=10000, n_jobs=-2)
optimal_lasso.fit(X_train, y_train)
print(optimal_lasso.alpha_)
```

```python
# look at cross-validation scores with this Lasso model
lasso = Lasso(alpha=optimal_lasso.alpha_)
lasso.fit(X_train, y_train)
lasso_scores = cross_val_score(lasso, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', lasso_scores.mean())
print('Training score: ', lasso.score(X_train, y_train))
print('Test score: ', lasso.score(X_test, y_test))
```

```python
print('MAE on training set: ', mean_absolute_error(y_train, lasso.predict(X_train)))
print('MAE on test set: ', mean_absolute_error(y_test, lasso.predict(X_test)))
```

### Random Forest

#### With power transform

```python
random_forest_pt = TransformedTargetRegressor(regressor=RandomForestRegressor(n_estimators=300,
                                                                              n_jobs=-2,
                                                                              random_state=1),
                                              transformer=PowerTransformer(method='box-cox'))
random_forest_pt.fit(X_train, y_train)
random_forest_pt_scores = cross_val_score(random_forest_pt, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', random_forest_pt_scores.mean())
print('Training score: ', random_forest_pt.score(X_train, y_train))
print('Test score: ', random_forest_pt.score(X_test, y_test))
```

#### Without power transform

```python
random_forest = RandomForestRegressor(n_estimators=200,
                                      min_samples_leaf=2,
                                      n_jobs=-2,
                                      random_state=1)

random_forest.fit(X_train, y_train)
random_forest_scores = cross_val_score(random_forest, X_train, y_train, cv=5)

print('Mean cross-validated training score: ', random_forest_scores.mean())
print('Training score: ', random_forest.score(X_train, y_train))
print('Test score: ', random_forest.score(X_test, y_test))
```

```python
print('MAE on training set: ', mean_absolute_error(y_train, random_forest.predict(X_train)))
print('MAE on test set: ', mean_absolute_error(y_test, random_forest.predict(X_test)))
```

```python
random_forest = RandomForestRegressor(random_state=1)

params = {'n_estimators': [200, 300, 400],
          'max_depth': [None],
          'min_samples_leaf': [1, 2], 
          'ccp_alpha': [0, 0.0001, 0.001]}

random_forest_gs = GridSearchCV(random_forest,
                                params,
                                n_jobs=-2, 
                                cv=5, 
                                verbose=3)

random_forest_gs.fit(X_train, y_train)
```

```python
print('Best parameters:')
print(random_forest_gs.best_params_)
print('Best estimator mean cross-validated training score:')
print(random_forest_gs.best_score_)
print('Best estimator score on the full training set:')
print(random_forest_gs.score(X_train, y_train))
print('Best estimator score on the test set:')
print(random_forest_gs.score(X_test, y_test))
```

```python
random_forest_gs_best = random_forest_gs.best_estimator_
```

```python
print('MAE on training set: ', mean_absolute_error(y_train, random_forest_gs_best.predict(X_train)))
print('MAE on test set: ', mean_absolute_error(y_test, random_forest_gs_best.predict(X_test)))
```
